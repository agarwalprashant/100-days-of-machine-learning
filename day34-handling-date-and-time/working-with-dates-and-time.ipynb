# Working with Dates

Jab date column ko object type mein store kiya jata hai, toh Python (specifically Pandas) usko ek string ki tarah treat karta hai. Iska matlab hai ki Pandas ko pata nahi hota ke us column ke andar actual dates hain, balki woh bas characters (strings) ko store kar raha hota hai.

Date ko object se convert karne ka fayda:

Jab tak tum us date column ko datetime format mein convert nahi karoge, tab tak Pandas ya koi bhi library uss date ke upar operations perform nahi kar payegi, jaise ki year, month, day nikaalna.
Solution: Tumhe pd.to_datetime() function ka use karna padega taaki Pandas ko samajh aaye ki ye column ek proper date format mein hai, jisse tum uske different components (jaise year, month, day) ko easily extract kar pao.

For example:

python
Copy code
df['date_column'] = pd.to_datetime(df['date_column'])
Iske baad tum ye operations kar paoge:

Year nikaalne ke liye: df['date_column'].dt.year
Month nikaalne ke liye: df['date_column'].dt.month
Day nikaalne ke liye: df['date_column'].dt.day
Toh, jab tak tum date ko string ya object ke form mein rakhoge, Pandas ke paas woh power nahi hoti usko analyze karne ki, isliye convert karna zaroori hai!

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#### 1. Extract year

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

#### 2. Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

#### Extract Days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```
